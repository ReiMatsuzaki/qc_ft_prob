```python
from scipy.stats import norm
mu, sigma = 3, 1
D_max = 5
Q_max = 5
demand_pdf = lambda x: norm.pdf(x, mu, sigma)
result = solve_newsvendor_qaoa(
    demand_dist=demand_pdf,
    c=1.0,
    lam=5.0,
    Q_max=Q_max,
    D_max=D_max,
    p=3,            # QAOA深さ
    M=2,            # フーリエ打ち切り次数
    n_shots=500,    # 測定回数
    verbose=True
)
```

```python
comprehensive_analysis(result)
```

```python
visualize_circuit_from_result(result)
```
